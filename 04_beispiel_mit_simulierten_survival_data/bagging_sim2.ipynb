### Weibull Modell Evaluation

```python
def ipc_weighted_mse(y_true, y_pred, sample_weight):
    return np.average((y_true - y_pred) ** 2, weights=sample_weight)

y_pred = aft.predict_survival_function(df_test.drop(['weights_ipcw', 'survived','time','event'], axis=1), times = tau).iloc[0].tolist()
weibull_mse = ipc_weighted_mse(y_true=df_test['survived'].values, y_pred=y_pred, sample_weight=df_test['weights_ipcw'])
print(f'weibull_ipcw_mse auf test data: {round(weibull_mse,4)}')

```

```python
def get_Nbi(lists):
    arr = np.array(lists)
    max_value = arr.max()
    counts = np.apply_along_axis(lambda x: np.bincount(x, minlength=max_value + 1), axis=1, arr=arr)
    return counts

# Beispiel für die Verwendung
x = [[0, 1, 1, 4, 5, 4, 4, 7], [0, 1, 1, 4, 5, 4, 4, 6]]
print(len(x))

c = get_Nbi(x)
print(len(c))
print(c)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Beispiel-Daten als pandas DataFrame
X_train = pd.DataFrame([[1, 2], [3, 4], [5, 6], [7, 8]], columns=["feature1", "feature2"])
y_train = [0, 1, 0, 1]

# Modell erstellen und trainieren
clf = RandomForestClassifier(n_estimators=100, bootstrap=True, random_state=42)
clf.fit(X_train, y_train)

# Beispiel-Testdatenpunkt als pandas DataFrame
X_erwartung = pd.DataFrame([[3, 4]], columns=["feature1", "feature2"])

# Vorhersagen aller Bäume im Wald erhalten (konvertiert in numpy-Array)
tree_predictions = np.array([tree.predict(X_erwartung.values) for tree in clf.estimators_])

# Durchschnittliche Vorhersage der einzelnen Bäume berechnen
average_prediction = tree_predictions.mean()

# Wahrscheinlichkeiten mit predict_proba berechnen
proba_prediction = clf.predict_proba(X_erwartung)[:, 1]

print("Durchschnittliche Vorhersage der einzelnen Bäume:", average_prediction)
print("Vorhersagewahrscheinlichkeit von clf.predict_proba:", proba_prediction)

X_erwartung.values
```
