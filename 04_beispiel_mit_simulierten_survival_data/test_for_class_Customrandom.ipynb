## Description of the Test:

The purpose of this test is to compare the performance and consistency of a custom implementation of a `RandomForestClassifier` (named `CustomRandomForestClassifier`) with the standard `RandomForestClassifier` from Scikit-learn.

### The test does the following:

1. **Data Generation**: 
   A synthetic dataset with 1000 samples and 20 features is generated, with 15 informative features and 2 classes.
   
2. **Model Training**:
   - A `RandomForestClassifier` with 2000 decision trees is trained on the training dataset.
   - The bootstrap samples used for each tree in the `RandomForestClassifier` are extracted.

3. **Custom Model Training**:
   - A `CustomRandomForestClassifier` is trained using the same number of trees, bootstrap samples, and hyperparameters as the `RandomForestClassifier`.

4. **Prediction Comparison**:
   - The predictions on the test dataset are compared between the two models in terms of accuracy and the percentage of agreement between their predictions.


```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from class_CustomRandomForestClassifier import CustomRandomForestClassifier

def test_custom_vs_standard_forest_large():
    # Generate a large dataset with 1000 samples and 20 features
    X_large, y_large = make_classification(n_samples=1000, n_features=20, n_informative=15, n_classes=2, random_state=42)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_large, y_large, test_size=0.3, random_state=42)

    # Step 1: Train the standard RandomForestClassifier
    rf = RandomForestClassifier(n_estimators=2000, random_state=42, bootstrap=True,
                                max_depth=3, max_features='sqrt', min_samples_split=2)
    rf.fit(X_train, y_train)

    # Get the predictions of the RandomForestClassifier on the test set
    rf_predictions = rf.predict(X_test)

    # Extract the bootstrap samples used by each tree
    bootstrap_samples = rf.estimators_samples_

    # Step 2: Train the CustomRandomForestClassifier using the same bootstrap samples
    custom_rf = CustomRandomForestClassifier(custom_samples=bootstrap_samples, n_estimators=2000, random_state=42,
                                             max_depth=3, max_features='sqrt', min_samples_split=2)
    custom_rf.fit(X_train, y_train)

    # Get the predictions of the CustomRandomForestClassifier on the test set
    custom_rf_predictions = custom_rf.predict(X_test)

    # Step 3: Compare the predictions and performance on the test data
    accuracy_rf = accuracy_score(y_test, rf_predictions)
    accuracy_custom_rf = accuracy_score(y_test, custom_rf_predictions)

    # Print the accuracy of both models
    print(f"Accuracy of RandomForestClassifier: {accuracy_rf:.4f}")
    print(f"Accuracy of CustomRandomForestClassifier: {accuracy_custom_rf:.4f}")

    # Check the percentage of matching predictions between the two models
    match_percentage = (rf_predictions == custom_rf_predictions).mean() * 100
    print(f"Prediction match percentage: {match_percentage:.2f}%")

# Run the test
test_custom_vs_standard_forest_large()
```
